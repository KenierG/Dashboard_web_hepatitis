```python
df = pd.read_csv('/content/drive/MyDrive/diplomado unicesmag/hepatitis.csv', header=0)
```

```python

for index, row in df.iterrows():
    if row['age'] <= 17:
        df.loc[index, 'age_category'] = 'young'
    elif row['age'] <= 60:
        df.loc[index, 'age_category'] = 'middle-aged'
    else:
        df.loc[index, 'age_category'] = 'elderly'
```

```python
age_categories = df['age_category'].value_counts()
print(age_categories)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
# Seleccionar características (variables independientes)
X = df[['steroid',	'antivirals',		'bilirubin',	'alk_phosphate',	'sgot'	,'albumin'	,'protime','fatigue'	,'malaise',	'anorexia',	'liver_big',	'liver_firm'	,'spleen_palable'	,'spiders',	'ascites'	,'varices',	'histology','sex']]
Y = df[['age_category']]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=24)
```

```python
modelo = RandomForestClassifier(n_estimators=100)
modelo.fit(X_train, Y_train)
```

```python
probabilidades = modelo.predict_proba(X_test)
```
